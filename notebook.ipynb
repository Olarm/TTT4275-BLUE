```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F_s = 10e6
T = 10e-6
f_0 = 10e5
w_0 = 2 * np.pi * f_0
phi = np.pi / 8
A = 1
```

```python
N = 513
n = np.arange(N)
x = np.zeros(N)
mu, sigma = 0, 1
v = np.random.normal(mu, sigma, N)

log_x = 1j * (w_0 * n * T + phi + v)
x = A * np.exp(log_x)

print(x.shape)

plt.figure(figsize=(12,6))
plt.plot(n, x)
plt.show()
```
